```python
def mounthly_mean(df):
    #Selcionando por mes de cada ano
    #A media mensal das acoes durante a pandemia
    Ano=[]
    Mes=[]
    Media_mglu=[]
    Media_viia=[]
    Media_amer=[]
    indice=[]

    for year in df.index.year.unique():
        for month in df.index.month.unique():

            media1= df[(df.index.year==year)&(df.index.month==month)].mean()["MGLU3.SA"]
            media2= df[(df.index.year==year)&(df.index.month==month)].mean()["VIIA3.SA"]
            media3= df[(df.index.year==year)&(df.index.month==month)].mean()["AMER3.SA"]

            Ano.append(year)
            Mes.append(month)
            Media_mglu.append(media1)
            Media_viia.append(media2)
            Media_amer.append(media3)
    df_mensal= pd.DataFrame(
        {'Ano': Ano,
         'Mes': Mes,
         'Magazine Luiza': Media_mglu,
         'Via Varejo': Media_viia,
         'Americanas': Media_amer,
        })
    for i in range(df_mensal.shape[0]):
        indice.append(datetime.date(df_mensal['Ano'][i],df_mensal['Mes'][i],1)) 
    df_mensal['Indice']= indice
    df_mensal.set_index('Indice', inplace=True)
    df_mensal.drop(['Ano','Mes'], axis=1, inplace=True)
    return df_mensal
```

```python
def monthly_mean(df):
    #Selcionando por mes de cada ano
    #A media mensal das acoes durante a pandemia
    Ano=[]
    Mes=[]
    Media_df=[]
    indice=[]

    for year in df.index.year.unique():
        for month in df.index.month.unique():

            media= df[(df.index.year==year)&(df.index.month==month)]['Close'].mean()

            Ano.append(year)
            Mes.append(month)
            Media_df.append(media)
    df_mensal= pd.DataFrame(
        {'Ano': Ano,
         'Mes': Mes,
         'Media': Media_df,
        })
    for i in range(df_mensal.shape[0]):
        indice.append(datetime.date(df_mensal['Ano'][i],df_mensal['Mes'][i],1)) 
    df_mensal['Indice']= indice
    df_mensal.set_index('Indice', inplace=True)
    df_mensal.drop(['Ano','Mes'], axis=1, inplace=True)
    return df_mensal
```
